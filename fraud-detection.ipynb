```python
def make_X(fraud_data):
    from sklearn.preprocessing import StandardScaler
    X = fraud_data.drop(['Class', 'Time'], axis=1)
    X['Amount'] = StandardScaler().fit_transform(X['Amount'].values.reshape(-1, 1))
    y = fraud_data['Class']
    return X, y
```

```python

X, y = make_X(fraud_data)
```
